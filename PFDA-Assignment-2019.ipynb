The above histogram plots the 100,000 random numbers generated using the randn function. As can be seen the histogram is bell shaped indicating visually that the numbers gnerated are in a normal/gaussian distribution. The number of bins has been set to 40 as 10 bins smoothes out the plot and does not give an accurate visualisation of the data generated.

In a normally distrbuted sample of data like this one the tallest peak in the KDE line of the plot indicates where the mean of the data lies. This supports the parameters that we gave in python as visually the tallest peak appears to be 10 which is what was initially set in generation of the data.

### Shuffle Function

Code for shuffle adapted from [here](https://docs.scipy.org/doc/numpy-1.16.0/reference/generated/numpy.random.shuffle.html#numpy.random.shuffle).

```python
# Seed set to 35
np.random.seed(35)

# Generate array for comparisson against shuffle function
np.random.rand(4,3)
```

```python
# Seed set to 35
np.random.seed(35)

# Create array with two column and four rows
df = np.random.rand(4,3)

# Shuffle function 
np.random.shuffle(df)
print(df)
```

The shuffle function as demonstrated in the above code takes an array, in this example an array with 3 columns and 4 rows of data and changes the order of the data on the fist axis of the array. Note that the shuffle function shuffles the array in place so the order of df changes.

As the seed was set at 35, the output of the array will be the same each time the code is run in the code that does not shuffle the data. The shuffle function then takes this data still using the seed of 35 and mixes up the data on the first axis with all subsequent axis following the row that they relate to from the first axis.

This can easily be seen by looking at the output of both the shuffled and unshuffled outputs, in that all the elements in row 1 of the unshuffled data has now moved to row 4 in the shuffled data.

I found it interesting that no matter how many times I ran the code the order of the shuffled data remained the same, my initial thought was that each time I would run the shuffle function that the data would appear in a different order. In practice however the order remained the same so there is obviously some sort of pre-defined order or seed that causes the shuffle order to remain the same even after re-running.

Initially I did not appreciate the reason for shuffling data like this, however now I understand it's purpose and how it can show how your analysis is a robust method in that taking random samples from the data gives the same result or generates a similar output.

### Permutation Function Array

Code for permutation adapted from [here](https://docs.scipy.org/doc/numpy-1.16.0/reference/generated/numpy.random.permutation.html#numpy.random.permutation).

```python
# Seed set to 35
np.random.seed(35)

# Generate array for comparisson against shuffle function
np.random.rand(4,3)
```

```python
# Seed set to 35
np.random.seed(35)

# Create array with two column and four rows
df = np.random.rand(4,3)

# Permiate df
np.random.permutation(df)

print(df)
```

```python
# Seed set to 35
np.random.seed(35)

# Permutation function with array
perm = np.random.permutation(df)

print(perm)
```

The permutations function differs in two ways from the shuffle function, I will explain the first difference with the aid of the 3 lines of code above.

The first row of code displays a multidimensional array with 3 columns and 4 rows with a seed of 35 using the random function displaying the output. 

The second line of code creates the same array and saves it as variable df, executes the permutation function and then prints df to output.

The third and final line of code creates a new variable (perm) again executing the permutations function and prints the variable to output.

Where permutation differs from shuffle is that permutation does not change the original df variable which can be seen in the output from first and second rows of code as the output is the same indicating that permutation did not make changes with the variable df. In order to see the interaction of permutation with df a new variable (perm) had to be created and printed to the output, this clealy illustrates a change in order of the df as the order of the rows has changed.

### Permutation Function Integer

```python
# Permutation function with integer
perm = np.random.permutation(9)
print(perm)
```

The second difference between the permutations function and the shuffle function occurs when an integer is passed into the permutations function rather than an array.

As an integer does not have a length an error will occur when passing an integer into the shuffle function, whereas with the permutations function when passed an integer the output of the code will be a range the length of the integer passed in a random order starting from 0 and up to but not including the integer passed.

This can be seen in the code above with a range of length 9 being the output ending on number 8 and ordered randomly.

### Normal Distribution

Code to draw vertical lines for standard deviations on plot adapted from [here](https://stackoverflow.com/questions/24988448/how-to-draw-vertical-lines-on-a-given-plot-in-matplotlib), code to hide certain elements of legend adapted from [here](https://discuss.analyticsvidhya.com/t/how-to-exclude-the-elements-from-the-legend-in-python/5393).

```python
# Seed set to 35
np.random.seed(35)

avg = 177
sd = 7.42

# Create array using random.normal function, simulates avg height of Irish Males (mean=177cm, sd=7.42cm)
df = np.random.normal(loc=177, scale=7.42, size=100000)

# Plot using Seaborn
plt.rcParams["figure.figsize"] = (15,7.5)

# Plot Histogram
sbn.distplot(df, color='blue')

# Mean
xposition = [avg]
for xc in xposition:
    plt.axvline(x=xc, color='k', linestyle='--',label='μ')

# 1 Standard Deviation
xposition = [avg-sd]
for xc in xposition:
    plt.axvline(x=xc, color='red', linestyle='--',label='+/-1 σ')

xposition = [avg+sd]
for xc in xposition:
    plt.axvline(x=xc, color='red', linestyle='--',label='_nolegend_')  
    
# 2 Standard Deviations
xposition = [avg-(sd*2)]
for xc in xposition:
    plt.axvline(x=xc, color='green', linestyle='--',label='+/-2 σ')

xposition = [avg+(sd*2)]
for xc in xposition:
    plt.axvline(x=xc, color='green', linestyle='--',label='_nolegend_')
    
# 3 Standard Deviations
xposition = [avg-(sd*3)]
for xc in xposition:
    plt.axvline(x=xc, color='darkviolet', linestyle='--',label='+/-3 σ')
    
xposition = [avg+(sd*3)]
for xc in xposition:
    plt.axvline(x=xc, color='darkviolet', linestyle='--',label='_nolegend_')
    
    
sbn.set_style("darkgrid")
plt.title('Fig 3:Simulation Normal Distribution Irish Male Height (cm)')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.legend()

# Save image to be used in readme.md
plt.savefig('Resources/Random_Number_Normal_Distribution_Hist_2.png')
plt.show()
```

The above histogram simuates the average height of Irish males in cm taken from tall.life as referenced in the readme file, teh numbers were generated using the random.normal function with the mean height set to 177cm and standard deviation of 7.42cm accross 100,000 individuals.

I felt rather than generating a random array of numbers that it would be more easier understood using more meaningful data like average height of the Irish Male Population.

### Binomial Distribution

```python
# Seed set to 35
np.random.seed(35)

# Number of times to repeat test
repeat = 100000

# Number of flips in experiment
flips = 10

# Create array using random.binomial function, simulates results of 10 coin flips repeated 100,000 times
df = np.random.binomial(flips, 0.5, size=repeat)
              
# Plot using Seaborn
plt.rcParams["figure.figsize"] = (15,7.5)

sbn.distplot(df, color='orange')
sbn.set_style("darkgrid")
plt.title('Fig 4:Simulation Binomial Distribution Coin Flip')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')

# Save image to be used in readme.md
plt.savefig('Resources/Random_Number_Binomial_Distribution_Hist.png')
plt.show()
```

The above histogram uses the random.binomial function to create a distribution and plot the data using seaborn. In my example I am simulating a coin being flipped 3 times giving a probability of heads of 50% per flip, this data is generated 100,000 times with the resulting number of heads being plotted onto the histogram.

This data is very powerful as it can be used to determine whatever number of floips you like by changing the flips variable, I will be using 10 flips as my sample and below I have adapted some code that will determine the probability of the number of heads as you enter occuring.

Code to count exact number of occurences in ndarray adapted from [here](https://stackoverflow.com/questions/28663856/how-to-count-the-occurrence-of-certain-item-in-an-ndarray-in-python).

The code below takes the array generated above and uses it to determine the chance of exactly 5 heads occuring in the 10 flips, this can be adjusted to determine any number of heads in any number of flips by changing the value variable below and the flips value above.

```python
# Number of heads to check
value = 5

# Number times that result occured
num_occur = np.count_nonzero(df == value)

# Total number of Attempts
num_attem = len(df)

# % chance of result occuring
result = round(((num_occur/num_attem)*100),2)


# Print result
print("There is a "+str(result)+"% chance that exactly "+str(value)+" head(s) will occur out of "+str(flips)+" flips when attempted "+str(repeat)+" times.")
```

### Poisson Distribution

Code to hide histogram bars adapted from [here](https://seaborn.pydata.org/generated/seaborn.distplot.html), code to add point on kde line adapted from [here](https://stackoverflow.com/questions/43565892/python-seaborn-distplot-y-value-corresponding-to-a-given-x-value), code to set scale and ticks of x-axis adapted from [here](https://stackoverflow.com/questions/54822884/how-to-change-the-x-axis-range-in-seaborn-in-python), code to add labels to points adapted from [here](https://stackoverflow.com/questions/14432557/matplotlib-scatter-plot-with-different-text-at-each-data-point).

Code to round values in ndarray adapted from [here](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ndarray.round.html).

```python
# Seed set to 35
np.random.seed(35)

# Create array using random.poisson function, simulates number of meteors per hour example from references
df = np.random.poisson(lam=5, size=1000)
              
# Plot using Seaborn
plt.rcParams["figure.figsize"] = (15,7.5)

ax = sbn.kdeplot(df, color='salmon')
sbn.set_style("darkgrid")
plt.title('Fig 5:Simulation Poisson Distribution Meteors per Hour')
plt.xlabel('Number of Meteors')
plt.ylabel('Probability')

# Points for KDE Line
data_x, data_y = ax.lines[0].get_data()

xi = range(0,13) # coordinate where to find the value of kde curve
yi = np.interp(xi,data_x, data_y)
ax.plot([xi],[yi], marker='o',color='blue')

# Set scale and ticks of x axis
ax.set_xlim(-1,13)
ax.set_xticks(range(0,13))

# Round array to 5 decimals
yi = np.ndarray.round(yi,2)

# Annotate markers
for i,txt in enumerate((yi)):
    ax.annotate(txt, (xi[i], yi[i]))

# Save image to be used in readme.md
plt.savefig('Resources/Random_Number_Poisson_Distribution_Hist.png')
plt.show()
```
