### Importing of Python Packages for Analysis

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sbn
```

Below code sets initial value of seed variable to 35, adapted from [here](https://subscription.packtpub.com/book/big_data_and_business_intelligence/9781785285110/2/ch02lvl1sec16/numpy-random-numbers).

At it's simplest the random function can be used to generate a simple array of random numbers (between 0 and 1 not including 1) with array dimensions provided by user from a uniform distribution, whereby the first number determines the number of rows and the second the number of columns. 

The uniform distribution simply means that each number (between 0 and 1 not including 1) has an equal chance of being generated, this can be further explained using plots as can be seen below. 

The seed is set using np.random.seed(number), this sets initial value of the seed which is used to generate the numbers, adapted from [here](https://subscription.packtpub.com/book/big_data_and_business_intelligence/9781785285110/2/ch02lvl1sec16/numpy-random-numbers) allowing for the same random numbers to be generated each time.


Code adapted from [here](https://docs.scipy.org/doc/numpy-1.16.0/reference/generated/numpy.random.rand.html#numpy.random.rand).

```python
# A simple dataset generated using the random function
# As can be seen first value determins the number of rows and second the number of columns

# Seed set to 35
np.random.seed(35)

np.random.rand(3,4)
```

```python
# Below code used to format plot info

font = {'weight' : 'bold',
        'size'   : 14}

plt.rc('font', **font)
```

Below Histogram displays the distribution of random numbers generated using rand function in Numpy (uniform distribution). Code adapted from [here](http://python-graph-gallery.com/histogram/) to generate seaborn plot. 


Size of plot adjusted using code from [here](https://www.mikulskibartosz.name/how-to-change-plot-size-in-jupyter-notebook/), styling of plot area code adapted from [here](https://seaborn.pydata.org/tutorial/aesthetics.html).

```python
# Seed set to 35
np.random.seed(35)

# Create dataframe with one column and 100,000 rows (Uniform Distribution)
df = np.random.rand(100000,1)

# Plot using Seaborn
plt.rcParams["figure.figsize"] = (15,7.5)

sbn.distplot(df, color='orange',  bins=10, kde=False)
sbn.set_style("darkgrid")
plt.title('Fig 1:Random Number Uniform Distribution Histogram')
plt.xlabel('Random Number')
plt.ylabel('Frequency')

# Save image to be used in readme.md
plt.savefig('Resources/Random_Number_Uniform_Distribution_Hist.png')
plt.show()
```

The above histogram plots the 100,000 random numbers generated using the rand function. As can be seen the histogram is rectangular in shape and each interval has an occurence of **~10,000**. The number of bins has been set to 10 meaning that 10 intervals had been created increasing in increments of 0.1. 

This is a typical uniform distribution as each number has an equal chance of being generated.

```python
# Seed set to 35
np.random.seed(35)

# Create dataframe with one column and 100,000 rows (Uniform Distribution)
df = np.random.randn(100000,1)+10

# Plot using Seaborn
plt.rcParams["figure.figsize"] = (15,7.5)

sbn.distplot(df, color='orange',  bins=40)
sbn.set_style("darkgrid")
plt.title('Fig 1:Random Number Normal Distribution Histogram')
plt.xlabel('Random Number')
plt.ylabel('Frequency')

# Save image to be used in readme.md
plt.savefig('Resources/Random_Number_Normal_Distribution_Hist.png')
plt.show()
```
